```python
import pandas as pd
```

```python
from google.cloud import storage
import os
import io
client = storage.Client()
bucket = client.get_bucket('nlp_final_data')

#blob = bucket.blob('data_no_stop.csv')
blob = bucket.blob('df_nlp.csv')
content = blob.download_as_string()

df_nlp = pd.read_csv(io.BytesIO(content))
```

```python
#blob = bucket.blob('data_no_stop.csv')
blob = bucket.blob('song_data.csv')
content = blob.download_as_string()

song_data = pd.read_csv(io.BytesIO(content))
```

```python
song_data = song_data[['track_name','artists','track_genre']].drop_duplicates(subset=['track_name','artists'])
```

```python
merged_df = df_nlp.merge(song_data, on=['track_name','artists'], how='inner')
```
